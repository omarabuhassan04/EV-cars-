# Impormt Librarys

```python
import pandas as pd
import joblib
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
```

# Read Data

```python
df = pd.read_csv("ev_charging_dataset_Final(in).csv")
df
```

# Encoding

```python
df_encoded = df.copy()


mapping_dict = {}


for col in df_encoded.columns:
    if df_encoded[col].dtype == object:
        unique_values = df_encoded[col].unique()
        col_map = {value: idx for idx, value in enumerate(unique_values)}
        mapping_dict[col] = col_map
        df_encoded[col] = df_encoded[col].map(col_map)
for col, mp in mapping_dict.items():
    print(f"\n {col}")
    print(mp)
```

```python
df_encoded
```

# cost

```python
X_cost = df_encoded.drop("Charging Cost (USD)", axis=1)
y_cost = df_encoded["Charging Cost (USD)"]

X_cost_train, X_cost_test, y_cost_train, y_cost_test = train_test_split(X_cost, y_cost, test_size=0.2, random_state=42)

joblib.dump((X_cost_train, X_cost_test, y_cost_train, y_cost_test),"data/cost_train_test_split.pkl")
```

# Time

```python
X_time = df_encoded.drop("Charging Time Difference (minutes)", axis=1)
y_time = df_encoded["Charging Time Difference (minutes)"]

X_time_train, X_time_test, y_time_train, y_time_test = train_test_split(X_time, y_time, test_size=0.2, random_state=42)

joblib.dump((X_time_train, X_time_test, y_time_train, y_time_test),"data/time_train_test_split.pkl")
```

# Long

```python
X_long = df_encoded.drop("Long Session", axis=1)
y_long = df_encoded["Long Session"]

X_long_train, X_long_test, y_long_train, y_long_test = train_test_split(X_long, y_long, test_size=0.2, random_state=42)

joblib.dump((X_long_train, X_long_test, y_long_train, y_long_test),"data/long_train_test_split.pkl")
```
